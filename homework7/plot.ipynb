```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
asd = np.loadtxt('matmul_times.txt')
```

```python
def linear(x,  a, b):
    return a*x + b

from scipy.optimize import curve_fit
```

```python
popt, pcov = curve_fit(linear, np.log(asd[:,0]), np.log(asd[:,1]))
```

```python
errors = np.sqrt(np.diag(pcov))
print(popt[0], " +/- ", errors[0])
```

```python
plt.plot(np.log(asd[:,0]), np.log(asd[:,1]), 'o')
plt.plot(np.log(asd[:,0]), linear(np.log(asd[:,0]),*popt), '-', c='Crimson', label=r'$T \sim N^\alpha$, $\alpha = 2.93 \pm 0.02$')
plt.xlabel("$\log N$")
plt.legend(loc='best')
plt.ylabel("$\log(t)$")
```
